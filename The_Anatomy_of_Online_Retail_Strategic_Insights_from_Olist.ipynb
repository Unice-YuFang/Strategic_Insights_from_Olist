**ER Diagram Explaination:**

This ER-diagram illustrates the relationships between the tables used in our analysis and their interconnections. We outlined the schema for each table and highlighted their relationships with corresponding foot notations.

Beginning with the two central tables, namely the "orders" and "orders_items" tables, both serve as pivotal connections to several other tables. The "orders" table establishes a one (optional) to many relationships with the "orders_review" table, signifying that an order can have none, one, or multiple reviews. Additionally, it maintains a one-to-one connection with the "orders_payment" table, as an order corresponds to only one payment type. Furthermore, it has a many-to-many connection with the "customer" table, given that a customer can place one or many orders.

The "orders_items" table establishes a many-to-one connection with the "orders" table, signifying that an order can have one or multiple items. It also forms a one-to-many connection with the "products" table, indicating that a product could be part of one or more orders. Furthermore, it has a one-to-many connection with the "sellers" table, as a seller can be associated with one or many orders.

The "sellers" table additionally maintains a one-to-optional connection with the "closed_deal" table, recognizing that a closed deal may or may not be on the seller's list. Moreover, the "closed_deal" table establishes an optional-to-one connection with the "marketing_qualified_leads" table, indicating that a marketing qualified lead may or may not lead to a closed deal.

These represent the tables utilized in our analysis and their respective relationships with one another.

### **7-1. Geolocation**

```python
%%bigquery
SELECT *
FROM `ba775-a09-fall23-project.olist_data.geolocation`
```

**Description:** The overview of the geolocation table.

```python
%%bigquery
SELECT
  DISTINCT geolocation_zip_code_prefix,
  geolocation_state
FROM `ba775-a09-fall23-project.olist_data.geolocation`
ORDER BY geolocation_zip_code_prefix;
```

**Description:** The total number of zip code prefixes is 19,023.

```python
%%bigquery
SELECT
  COUNT(distinct geolocation_city) AS city_count,
  geolocation_state
FROM `ba775-a09-fall23-project.olist_data.geolocation`
GROUP BY geolocation_state
ORDER BY geolocation_state DESC;
```

**Description:** There are 27 cities in the table.

```python
%%bigquery
SELECT
  DISTINCT geolocation_city,
  geolocation_lat,
  geolocation_lng
FROM `ba775-a09-fall23-project.olist_data.geolocation`
ORDER BY geolocation_city ;
```

**Description:** We've noticed inconsistency in the way city data is recorded; the same city might have multiple variations in its name (for instance: '4o. centenario' and '4º centenario' refer to the same city). Additionally, each city has multiple sets of coordinates. Since our analysis focuses on the smallest geographic unit, which is the city, we don’t require detailed coordinates for each city.

```python
%%bigquery
CREATE OR REPLACE TABLE olist_data.geo_clean AS
SELECT
  g1.geolocation_zip_code_prefix,
  g1.geo_state,
  g2.geo_city,
  g3.geo_latitude,
  g3.geo_longitude
FROM
  (SELECT
    geolocation_zip_code_prefix,
    MAX(geolocation_state) AS geo_state
  FROM
    `ba775-a09-fall23-project.olist_data.geolocation`
  GROUP BY
    geolocation_zip_code_prefix) AS g1
JOIN
  (SELECT
    geolocation_zip_code_prefix,
    MAX(geolocation_city) AS geo_city
  FROM
    `ba775-a09-fall23-project.olist_data.geolocation`
  GROUP BY
    geolocation_zip_code_prefix) AS g2
    ON g1.geolocation_zip_code_prefix = g2.geolocation_zip_code_prefix
JOIN
  (SELECT
    geolocation_zip_code_prefix,
    AVG(geolocation_lat) AS geo_latitude,
    AVG(geolocation_lng) AS geo_longitude
  FROM
    `ba775-a09-fall23-project.olist_data.geolocation`
  GROUP BY
    geolocation_zip_code_prefix) AS g3
    ON g1.geolocation_zip_code_prefix = g3.geolocation_zip_code_prefix;
```

**Description:** We took the most frequently occurring city and state names from the geolocation table for each zip code prefix. These were used as the city and state names for the respective zip code prefixes. Additionally, we averaged the latitude and longitude for each zip code prefix. This cleaned table was named 'geo_clean' to facilitate further analysis.

### **8-1. Users Analysis**

---


Starting with customers, we want to look at how often they come back to understand if it is a significant factor in determining the firms success. We're breaking down each small question and digging into the details to see how things are going. The goal is to find useful patterns that can help us make smart decisions for keeping customers engaged and happy.

**Question:** How many customers came back to the platform to make another purchase?

```python
%%bigquery

#New vs returning customers
WITH table1 AS(
  SELECT
    a.*,
    b.customer_unique_id
  FROM
    `ba775-a09-fall23-project.olist_data.orders` a
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.customers` b
  ON a.customer_id = b.customer_id) ,

table2 AS (
  SELECT
    customer_unique_id ,
    extract(year from order_purchase_timestamp) as order_year,
    DENSE_RANK() OVER (PARTITION BY customer_unique_id ORDER BY TIMESTAMP(order_purchase_timestamp)) as order_row_number
  FROM table1 ) ,

table3 AS (
  SELECT
    customer_unique_id,order_year,
    CASE WHEN order_row_number = 1 THEN 'new_customer' ELSE 'returning_customer' END as customer_type
  FROM
    table2)

SELECT
   order_year,
   COUNT(distinct case when customer_type = 'new_customer' then customer_unique_id end ) as New_cus,
   COUNT(distinct case when customer_type = 'returning_customer' then customer_unique_id end ) as returning_cus
FROM table3
GROUP BY
  order_year
ORDER BY
  order_year asc;
```

**Inference** : In 2016, the platform had a modest beginning with 326 new customers and only 3 returning customers. However, there was a significant increase in 2017, with new customers rising to 43,708 and returning customers to 1,116. This trend continued into 2018, with new customers reaching 52,062 and returning customers increasing to 1,681. The data indicates not only a rapid expansion of the customer base but also a growing rate of customer retention year-over-year.

It is also surprising that only few customers have placed multiple orders in the same year.

### **8-2. Orders Analysis**
---

Now that we understood customer trend, we further dive into the orders placed by the customers to comprehend the trends, to see how the e-commerce giant might have used the data to strategically deploy their marketing campaigns.

**Question:** What are the peak hours for customer orders?

```python
%%bigquery
SELECT
  EXTRACT(HOUR FROM order_purchase_timestamp) AS hour,
  COUNT(*) AS order_count
FROM
   `ba775-a09-fall23-project.olist_data.orders`
GROUP BY
   hour
ORDER BY
   order_count DESC;
```

**Inference:** We noticed a significant surge in online activity around 4 pm, and it seems to be linked to the end of the typical workday and school hours. This time frame appears to be a prime moment for a boost in online shopping, likely because people are finishing up their daily responsibilities, making it a high-demand period for grabbing things online. Similar analysis done from US data, from this [article](https://www.statista.com/statistics/476827/usa-peak-online-shopping-hours/) also backs up the credibility of the inference from this data set.

**Question:** What are the peak months for customer orders?

```python
%%bigquery
SELECT
  EXTRACT(MONTH FROM order_purchase_timestamp) AS month,
  COUNT(*) AS order_count
FROM
  `ba775-a09-fall23-project.olist_data.orders`
GROUP BY
  month
ORDER BY
  order_count DESC;
```

**Inference:** As August rolls around, schools and universities kick off a new academic year, and this has a big impact on online shopping. It is a shopping spree season for students and teachers, where everyone is on the lookout to buy supplies, books and high tech gadgets. So, online stores get pretty busy as people gear up for the school year ahead. We come to a conclusion that usually at 4pm and specifically during August, there is peak surge on the online shopping. Now with this information, we want to understand which are the top categories which are mostly ordered to narrow down to few products.

**Question:**  What are the top five categories that have the most orders ?

```python
%%bigquery
SELECT
  product_category_name_english AS product_category,
  count(order_id) as order_count
FROM (
  SELECT
    C.order_id,
    D.product_category_name,
    E.product_category_name_english
  FROM
    `ba775-a09-fall23-project.olist_data.order_items` as C
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.products` AS D
  ON
    C.product_id = D.product_id
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.category` AS E
  ON
    D.product_category_name = E.product_category_name)
GROUP BY
  product_category_name_english
ORDER BY
  order_count DESC;
```

**Inference** :The top five categories that have most orders are bed_bath_table, health_beauty, sports_leisure, furniture_decor and computers_accessories. We believe that these product categories contain products that people either use on a daily basis or they are home necessities. Also their prices are usually affordable by most families. Hence they are popular among customers.


### **8-3. Products Analysis**
---

As we understand the most ordered products, we also wanted to see what are the variables affecting these products to be most sought after.
One of the key features observed is the presence of photo for a product on sale, which is what we want to analyse further.

**Question:** Will customers buy more if the product has more photos?

We aim to see how many photos will be optimal to get attention of a customer to select and order the particualr product.

```python
%%bigquery
WITH
A AS (
  SELECT
    product_id,
    COUNT(*) AS n_purchased
  FROM
    `ba775-a09-fall23-project.olist_data.order_items`
  GROUP BY
    product_id),

B AS (
  SELECT
    product_id,
    COALESCE(product_photos_qty, 0) AS product_photos_qty
  FROM
   `ba775-a09-fall23-project.olist_data.products` )

SELECT
  ROUND(AVG(a.n_purchased),2) AS avg_n_purchased,
  B.product_photos_qty
FROM
  A
LEFT JOIN
  B
USING
  (product_id)
GROUP BY
  b.product_photos_qty
ORDER BY
  b.product_photos_qty DESC;
```

**Inference:** While shopping online, we usually consider the product's photos to determine if it aligns with our expectations. Therefore, we are interested in understanding whether more information provided by the seller increases the likelihood of attracting customers.

The results indicate that products published with seven photos have the highest average number of purchases (average number of purchases = 4.38). Additionally, we observe a slight trend between 0-7 photos, where the more photos published for the products, the higher the average number of purchases. However, with more than seven photos, there doesn't seem to be a clear correlation with higher purchases. Therefore, we recommend that sellers include about seven photos for their products to attract customers.

---


Now that analysis being over all, we wanted dive one step further to see what the trend is for each product category with respect to product categpry.

**Question:** What is the recommended number of photos for each product category?

```python
%%bigquery
WITH n_purchased AS (
  SELECT
    product_id,
    COUNT(*) AS n_purchased
  FROM
    `ba775-a09-fall23-project.olist_data.order_items`
  GROUP BY
    product_id),

product_photos_qty AS (
  SELECT
    product_id,
    COALESCE(product_photos_qty, 0) AS product_photos_qty
  FROM
    `ba775-a09-fall23-project.olist_data.products`),

best_n_photo AS (
  SELECT
    ROUND(AVG(A.n_purchased), 2) AS avg_n_purchased,
    B.product_photos_qty,
    D.product_category_name_english,
    ROW_NUMBER() OVER (PARTITION BY D.product_category_name_english ORDER BY ROUND(AVG(A.n_purchased), 2) DESC) AS row_num
  FROM
    n_purchased AS A
  LEFT JOIN
    product_photos_qty AS B
    USING
      (product_id)
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.products` AS C
    ON B.product_id = C.product_id
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.category` AS D
    ON C.product_category_name = D.product_category_name
  WHERE
      D.product_category_name_english IS NOT NULL
  GROUP BY
      B.product_photos_qty, D.product_category_name_english)

SELECT
  avg_n_purchased,
  product_photos_qty,
  product_category_name_english
FROM
  best_n_photo
WHERE
  row_num = 1
ORDER BY
  product_photos_qty DESC;
```

**Inference:** As we divided all the products into different categories, we found varying recommended numbers of photos for each product category. The highest number of photos was recommended for products in the toys, art, and baby categories, where sellers are advised to provide 10 photos as this may result in the highest number of purchases. Following closely are the home_appliances_2, books_general_interest, and musical_instruments categories, for which it is recommended to have 9 photos to potentially achieve better numbers of purchases.

Another major factor determing the purchase of an order is the reviews that particular prodcut has. Looking further to see if more reviews or higher rating for an order is directly impacting a product's sale. The following questions aim to uncover those trends.

**Question:** Do people buy more if the product has more reviews or higher ratings?


```python
%%bigquery
WITH
product AS(
 SELECT
   order_id,
   product_id,
   SUM(price) AS total_price,
   COUNT(*) AS n_of_purchased
 FROM
   `ba775-a09-fall23-project.olist_data.order_items`
 GROUP BY
   order_id, product_id
 HAVING
   COUNT(DISTINCT product_id) = 1),

review AS (
 SELECT
   order_id,
   AVG(review_score) AS review_score,
   1 AS review
 FROM
   `ba775-a09-fall23-project.olist_data.order_review`
 GROUP BY
   order_id),

product_review AS (
 SELECT
   A.product_id,
   SUM(A.total_price) AS product_total_price,
   SUM(A.n_of_purchased) AS product_n_of_purchased,
   COUNT(*) AS product_n_of_order,
   AVG(B.review_score) AS product_review_score,
   sum(review) AS n_review
 FROM
   product AS A
 INNER JOIN
   review AS B
 USING
   (order_id)
 GROUP BY
   A.product_id)

SELECT
  CASE
    WHEN product_review_score > 0 AND product_review_score <= 1 THEN '(0, 1)'
    WHEN product_review_score > 1 AND product_review_score <= 2 THEN '(1, 2)'
    WHEN product_review_score > 2 AND product_review_score <= 3 THEN '(2, 3)'
    WHEN product_review_score > 3 AND product_review_score <= 4 THEN '(3, 4)'
    WHEN product_review_score > 4 AND product_review_score <= 5 THEN '(4, 5)'
    END AS review_score_range,
  COUNT(DISTINCT product_id) AS n_items,
  ROUND(AVG(product_n_of_purchased),3) AS avg_n_of_purchased,
  ROUND(AVG(product_n_of_order),3) AS avg_n_of_order,
  ROUND(AVG(n_review),3) AS avg_n_review,
  ROUND(AVG(product_total_price),3) AS avg_total_price
FROM
  product_review
GROUP BY
  review_score_range
ORDER BY
  review_score_range DESC;
```

**Inference** : People usually look at the satisfaction score shared by previous customers to assess the product's quality, influencing their decision on whether to make a purchase. Therefore, we used the data to examine the average number of purchases for each review score, aiming to determine if products with higher scores were more frequently purchased.
The results indicate that there were 18,786 products rated between 4 and 5, but the number of orders was only 3.26, which is lower than the 7,197 products with a satisfaction score between 3 and 4 that had the highest number of orders at 4.08. However, when compared to products with a satisfaction score lower than 3, the higher satisfaction score products did indeed have a higher number of orders.

### **8-4. Operations Analysis**
---

As we move from product to more on the operational front of the e-commerce giant, we have features like shipping and delivery time, to measure the purchase value and understand the dynamics of this business model.

**Question:** What proportion of the total item value is attributed to shipping expenses?

```python
%%bigquery
SELECT
  order_item_id,
  Count(*) as item_count,
  ROUND(AVG(price),2) as item_value,
  ROUND(AVG(freight_value),2) as freight_value,
  ROUND(AVG(price+freight_value),2) as total_value,
  FORMAT('%.2f%%', (AVG(freight_value) / AVG(price+freight_value) * 100)) AS freight_ratio
FROM
  `ba775-a09-fall23-project.olist_data.order_items`
GROUP BY
  order_item_id
ORDER BY
  order_item_id DESC ;
```

**Inference:** We're interested in determining the percentage of shipping costs relative to the total item value. Given that orders may vary in terms of the number of items, we've categorized orders with different item quantities into distinct groups. Upon analysis, we observed that 90% of the groups fall within the range of 10-20%. Notably, for order_item_id = 21, there is a 45% value, but this is based on only one occurrence and can be disregarded.

Delivery of a product involves shipping, along with one major factor of the delivery time. In this fast paced world, delivery time places a crucial role in determing a product's sale on multiple levels. This has been a challenge to most of the top firms to deliver at the earliest, and most of the business models work towards achieving minium delivery time.

**Question:** What's the average delivery time for each category when considering orders that have only a single item ?

```python
%%bigquery
WITH order_table AS (
  SELECT
    order_id,
    order_purchase_timestamp,
    order_delivered_customer_date,
    DATE_DIFF(order_delivered_customer_date,order_purchase_timestamp,day) as delivery_date,
    DATE_DIFF(order_delivered_customer_date,order_purchase_timestamp,hour) as delivery_hour,
    order_estimated_delivery_date,
    CASE WHEN DATE_DIFF(order_delivered_customer_date,order_estimated_delivery_date,day) < 0 THEN 'Early'
       WHEN DATE_DIFF(order_delivered_customer_date,order_estimated_delivery_date,day) > 0 THEN 'Late'
       ELSE 'On time' END AS delivery_status
  FROM
    `ba775-a09-fall23-project.olist_data.orders`
  WHERE
    order_purchase_timestamp IS NOT NULL
    AND order_delivered_customer_date IS NOT NULL
    AND order_estimated_delivery_date IS NOT NULL),

order_item_table AS (
  SELECT
    order_id, order_item_id, product_id
  FROM
    `ba775-a09-fall23-project.olist_data.order_items`
  WHERE
    order_id NOT IN (
      SELECT
        order_id
      FROM
        `ba775-a09-fall23-project.olist_data.order_items`
      GROUP BY
        order_id
      HAVING
        count(order_item_id) > 1)),

product_table AS (
  SELECT
    products_category.product_id, category_name.product_category_name_english
  FROM
    `ba775-a09-fall23-project.olist_data.products` AS products_category
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.category` AS category_name
  ON
    products_category.product_category_name = category_name.product_category_name
  WHERE
    products_category.product_category_name IS NOT NULL),

calcualte_table AS (
  SELECT
    X.order_id, X.order_purchase_timestamp, X.order_delivered_customer_date, X.delivery_date, X.delivery_hour, Y.product_id, Z.product_category_name_english as category
  FROM
    order_table AS X
  LEFT JOIN
    order_item_table AS Y
  ON X.order_id = Y.order_id
  LEFT JOIN
    product_table AS Z
  ON Y.product_id = Z.product_id
  WHERE
    Y.product_id IS NOT NULL and Z.product_category_name_english IS NOT NULL)

SELECT
  category, ROUND(AVG(delivery_date),0) AS average_delivery_day,
    ROUND(AVG(delivery_hour),1) AS average_delivery_hour
FROM
  calcualte_table
GROUP BY
  category
ORDER BY
  average_delivery_hour DESC ;
```

**Inference** : We can see that office_furniture takes a long time to deliver, which is reasonable since they tend to be in large pieces and might come from storing facilities at further distances. Fashion_shoes, christmas_supplies and security_and_services usually take about half a month, whereas the rest of the product categories (around 95%) arrive within a week or two, which still aren’t really fast compared to delivery services in America.

**Question:** After becoming a seller, the first order typically occurs within 86 days. Therefore, calculate the number of orders completed within approximately twice that duration. During the initial 180-day period after sellers join the platform, which product categories yield higher order volumes per seller?


```python
%%bigquery
SELECT
  product_category_name_english,
  COUNT(DISTINCT order_id) AS order_count,
  COUNT(DISTINCT seller_id) AS seller_count,
  ROUND(COUNT(DISTINCT order_id) / COUNT(DISTINCT seller_id),0) AS order_per_seller
FROM
  `ba775-a09-fall23-project.olist_data.order_items`
INNER JOIN
  `ba775-a09-fall23-project.olist_data.closed_deals`
  USING(seller_id)
INNER JOIN
  `ba775-a09-fall23-project.olist_data.orders`
  USING(order_id)
INNER JOIN
  `ba775-a09-fall23-project.olist_data.products`
  USING(product_id)
INNER JOIN
  `ba775-a09-fall23-project.olist_data.category`
  USING(product_category_name)
WHERE
  (TIMESTAMP_DIFF(order_purchase_timestamp, won_date, DAY)) <= 180
GROUP BY
  product_category_name_english
ORDER BY
  order_per_seller DESC;
```

**Inference:** During the initial 180-day period after becoming a seller, categories such as technical books, watches gifts, and audio exhibit the highest average number of orders per seller. This could be attributed to an increased demand for these products coupled with a relatively lower number of sellers, potentially granting these categories a market advantage, resulting in a higher volume of orders. However, despite the health beauty category having the highest overall number of product orders, the greater number of sellers within this category diminishes the average orders per seller in comparison to the top three. This scenario suggests a heightened demand for health beauty products, but also highlights a more competitive landscape among sellers within this category.

###8-6. Recency, Frequency, Monetary Value
---

As per Investopedia, Recency, Frequency, Monetary Value (RFM) is a powerful model in marketing analysis, categorizing a company's customer base based on their buying behaviors. It examines how recent a customer made a purchase, how frequently they buy, and the amount they spend. We see this as a valuable metric to grasp the intricate dynamics of our company. This section provides a concise overview of the RFM model and its significance in understanding our business landscape.

```python
%%bigquery

WITH
  table1 AS (
  SELECT
    a.*,
    b.customer_unique_id,
    c.payment_value
  FROM
    `ba775-a09-fall23-project.olist_data.orders` a
  LEFT JOIN
    `ba775-a09-fall23-project.olist_data.customers` b
  ON
    a.customer_id = b.customer_id
  LEFT JOIN (
    SELECT
      order_id,
      SUM(payment_value) AS payment_value
    FROM
      `ba775-a09-fall23-project.olist_data.order_payments`
    GROUP BY
      1) c
  ON
    a.order_id = c.order_id),
  table2 AS (
  SELECT
    customer_unique_id,
    DATE_DIFF(DATE '2018-11-01', MAX(DATE(order_purchase_timestamp)), DAY) AS Recency,
    COUNT(DISTINCT order_id) AS Frequency,
    SUM(payment_value) AS Monetary
  FROM
    table1
  GROUP BY
    1),
  table3 AS (
  SELECT
    *,
    NTILE(3) OVER (ORDER BY recency DESC) AS recency_Tertile,
    NTILE(3) OVER (ORDER BY frequency ) AS frequency_Tertile,
    NTILE(3) OVER (ORDER BY monetary ) AS monetary_Tertile
  FROM
    table2 ),
  table4 AS (
  SELECT
    *,
    recency_Tertile + frequency_Tertile + monetary_Tertile AS rfm_score,
    CAST(CONCAT(recency_Tertile, frequency_Tertile, monetary_Tertile) AS int) AS rfm
  FROM
    table3),
  table5 AS (
  SELECT
    *,
    CASE
      WHEN rfm IN (311, 312, 311) THEN 'new customers'
      WHEN rfm IN (111,
      121,
      131,
      122,
      133,
      113,
      112,
      132) THEN 'lost customers'
      WHEN rfm IN (212, 313, 123, 221, 211, 232) THEN 'regular customers'
      WHEN rfm IN (223,
      222,
      213,
      322,
      231,
      321,
      331) THEN 'loyal customers'
      WHEN rfm IN (333, 332, 323, 233) THEN 'champion customers'
  END
    rfm_segment
  FROM
    table4 )
  --select * from table5 where rfm_segment = 'new customers'
SELECT
  rfm_segment,
  COUNT(DISTINCT customer_unique_id) AS number_of_customers,
  AVG(recency) as Average_recency,
  AVG(frequency) as Average_frequency,
  AVG(monetary) as Average_monetary
FROM
  table5
GROUP BY
  rfm_segment
  order by Average_monetary desc;
```

From the above result, we see "champion customers" are the one with highest Monetary value, followed by the highest number of average orders. However "new customers" were the most recent customers and "regular customers" comprising of the major portion of the customer segment.
